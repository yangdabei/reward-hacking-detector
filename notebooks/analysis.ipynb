## 1. Load Experiment Results

```python
# Load sweep results if available
results_path = Path('../data/sweep_results.json')
if results_path.exists():
    with open(results_path) as f:
        sweep_results = json.load(f)
    print(f'Loaded {len(sweep_results)} sweep results')
else:
    print('No sweep results found. Run: python -m src.cli sweep')
    sweep_results = []
```

## 2. Training Curves

```python
# Plot training curves (episode reward over time)
# TODO: Load episode_rewards from your experiment results and plot here
fig, ax = plt.subplots()
ax.set_xlabel('Episode')
ax.set_ylabel('Total Reward')
ax.set_title('Training Curve')
ax.text(0.5, 0.5, 'Run experiments first', transform=ax.transAxes,
        ha='center', va='center', fontsize=14, color='gray')
plt.tight_layout()
plt.show()
```

## 3. Proxy Reliance vs Coin Reward

```python
# Plot: x-axis = coin reward, y-axis = proxy reliance score
# This reveals at which coin reward value hacking emerges
if sweep_results:
    coin_rewards = [r.get('coin_reward', 0) for r in sweep_results]
    proxy_scores = [r.get('proxy_reliance_score', 0) for r in sweep_results]
    fig, ax = plt.subplots()
    ax.scatter(coin_rewards, proxy_scores, alpha=0.6)
    ax.axhline(y=0.3, color='r', linestyle='--', label='Hacking threshold')
    ax.set_xlabel('Coin Reward')
    ax.set_ylabel('Proxy Reliance Score')
    ax.set_title('Reward Hacking Emergence')
    ax.legend()
    plt.tight_layout()
    plt.show()
else:
    print('No sweep results to plot')
```

## 5. Reward Decomposition

```python
# Bar chart: reward sources (goal, coin, step penalty, lava)
# Compare aligned agent vs hacking agent
labels = ['Goal', 'Coin', 'Step Penalty', 'Lava']
aligned = [10.0, 0.0, -1.5, 0.0]   # placeholder
hacking = [0.0, 5.0, -3.0, 0.0]    # placeholder

x = np.arange(len(labels))
width = 0.35
fig, ax = plt.subplots()
ax.bar(x - width/2, aligned, width, label='Aligned agent')
ax.bar(x + width/2, hacking, width, label='Hacking agent')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylabel('Mean Reward Contribution')
ax.set_title('Reward Decomposition (placeholder data)')
ax.legend()
plt.tight_layout()
plt.show()
```
